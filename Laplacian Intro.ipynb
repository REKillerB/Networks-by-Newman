```python
import networkx as nx
import numpy as np
G=nx.Graph()
G.add_edges_from([(1,2),(1,3),(1,4),(2,3),(3,4),(4,2),(2,5),(2,6),(3,5),(3,6),(4,5),(4,6),(5,7),(6,7)])
deg=[v*75 for k, v in G.degree]
nx.draw_kamada_kawai(G, node_size=deg, node_color=deg)
```

```python
degrees=nx.linalg.adjacency_matrix(G).todense().sum(axis=1)
D=np.zeros((degrees.shape[0], degrees.shape[0]))
for i in range(degrees.shape[0]):
    D[i,i]=degrees[i]
D
```

```python
A=nx.linalg.adj_matrix(G).todense()
A
```
